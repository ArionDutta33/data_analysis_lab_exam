```python
print("Checking for any missing values and summarizing data")
df=pd.read_csv("employee_survey.csv")
missing_values = df.isnull().sum()
data_types = df.dtypes
summary_stats = df.describe(include='all')
missing_values, data_types, summary_stats
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 3, figsize=(18, 5))

print("Age distribution analysis")
axs[0].hist(df['Age'], bins=15, color='skyblue', edgecolor='black')
axs[0].set_title('Age Distribution')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Frequency')

print("Experience distribution analysis")
axs[1].hist(df['Experience'], bins=15, color='orange', edgecolor='black')
axs[1].set_title('Experience Distribution')
axs[1].set_xlabel('Years of Experience')
axs[1].set_ylabel('Frequency')

print("Job satisfaction analysis")
axs[2].hist(df['JobSatisfaction'], bins=5, color='green', edgecolor='black')
axs[2].set_title('Job Satisfaction Distribution')
axs[2].set_xlabel('Satisfaction Level')
axs[2].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
